```python
#checking for the relation between diffrent variables i.e. attack and serve are corelated
#if they are greater than 0.6 they are positively related if -0.6 they are negatiely corelated like propotionality
numeric_cols = df.select_dtypes(include = ["int","float"]).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize= (12,10))
sns.heatmap(corr_matrix, annot= True, cmap="coolwarm", linewidth= 5)
plt.show()
```

```python
position_counts = df['Position'].value_counts()
```

```python
position_counts
```

```python
plt.pie(position_counts , labels=position_counts.index, startangle = 90, autopct='%1.1f%%')
plt.title("Positions in VolleyBall")
plt.show()
```

```python
avg_attack_by_country = df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False)
```

```python
avg_attack_by_country.sort_values(ascending=False).head(5).plot(kind="bar")
plt.title("Attack by country")
plt.xlabel("Country")
plt.ylabel("Attack")
plt.show()
```

```python
serve_trend_by_age = df.groupby("Age")["Serve"].mean()

serve_trend_by_age.plot(kind="line", marker="o", linestyle="-", color = "orange")
plt.title("Serve tren over age group")
plt.xlabel("Age")
plt.ylabel("Average Serve")
plt.show()
```
