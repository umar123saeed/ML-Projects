```python
import numpy as np
import cv2

image = cv2.imread('digits1.jpeg')
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# safer splitting
rows = np.vsplit(gray_img, 40)      # 👈 FIXED (was 50)
cells = [np.hsplit(r, 40) for r in rows]  # 👈 FIXED (was 100)

NP_array = np.array(cells)

print("NP_array shape:", NP_array.shape)

# -------------------
# reshape data
# -------------------
train_data = NP_array[:, :20].reshape(-1, 400).astype(np.float32)
test_data  = NP_array[:, 20:40].reshape(-1, 400).astype(np.float32)

print("train_data:", train_data.shape)
print("test_data:", test_data.shape)

# -------------------
# FIXED labels (MATCH 1600)
# -------------------
train_labels = np.repeat(np.arange(10), 160).reshape(-1, 1).astype(np.float32)
test_labels  = np.repeat(np.arange(10), 160).reshape(-1, 1).astype(np.float32)

print("train_labels:", train_labels.shape)

# safety check
assert train_data.shape[0] == train_labels.shape[0], "Mismatch fixed incorrectly!"

# -------------------
# KNN
# -------------------
knn = cv2.ml.KNearest_create()
knn.train(train_data, cv2.ml.ROW_SAMPLE, train_labels)

# prediction
ret, output, neighbours, distance = knn.findNearest(test_data, k=3)

accuracy = np.mean(output == test_labels) * 100

print("Accuracy:", accuracy)
```
